```python
import torch
import pandas as pd
import numpy as np
from torch import nn
import torch_directml
```

```python
gpu=torch_directml.device(0)
gpu
```

```python
df=pd.read_csv("ObesityDataSet_raw_and_data_sinthetic.csv")
df
```

```python
x=df.drop(["NObeyesdad"],axis=1)
y=df["NObeyesdad"]
```

```python
x.replace({"Female":0,"Male":1,"yes":1,"no":0},inplace=True)
```

```python
x
```

```python
x=pd.get_dummies(x)
x
```

```python
x.replace({True:1,False:0},inplace=True)
x
```

```python
y=pd.get_dummies(y)
y.replace({True:1,False:0},inplace=True)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train=torch.Tensor(np.array(x_train)).to(gpu)
y_train=torch.Tensor(np.array(y_train)).to(gpu)
x_test=torch.Tensor(np.array(x_test)).to(gpu)
y_test=torch.Tensor(np.array(y_test)).to(gpu)
```

```python
class ObesityClassifier(nn.Module):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.layer1=nn.Sequential(
            nn.Linear(in_features=26,out_features=128),
            nn.ReLU(),
            nn.Linear(128,32),
            nn.ReLU(),
            nn.Linear(32,7),
            nn.LogSoftmax(),
        )
    def forward(self,x):
        return self.layer1(x)
```

```python
model=ObesityClassifier().to(gpu)
```

```python
loss_fn=nn.CrossEntropyLoss()
opt=torch.optim.Adam(model.parameters(),lr=0.001)
```

```python
epochs=50
for epoch in range(epochs):
    model.train()
    y_preds=model.forward(x_train)
    loss=loss_fn(y_preds,y_train)
    opt.zero_grad()
    loss.backward()
    opt.step()
    print(loss)
```

```python
try:
    y_test=y_test.cpu()
    y_test=y_test.detach().numpy()
except:
    "error"
```

```python
with torch.inference_mode(mode=False):
    y_res=model.forward(x_test).cpu()
    y_res=y_res.detach().numpy()
    c=0
    for i in range(len(y_res)):
        if (y_res[i].argmax()==y_test[i].argmax()):
            c+=1
    print((c/len(y_res))*100)
```
