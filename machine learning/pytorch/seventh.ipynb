```python
import torch
from torch import nn
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Heart Prediction Quantum Dataset.csv")
```

```python
df
```

```python
x=df.drop(axis=1,columns=["HeartDisease"])
y=df["HeartDisease"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_test
```

```python
x_train=torch.Tensor(np.array(x_train))
y_train=torch.unsqueeze(torch.Tensor(np.array(y_train)),dim=1)
x_test=torch.Tensor(np.array(x_test))
y_test=torch.unsqueeze(torch.Tensor(np.array(y_test)),dim=1)
```

```python
y_test
```

```python
class DiseasePredict(nn.Module):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.layer1=nn.Sequential(
            nn.Linear(in_features=6,out_features=16),
            nn.Linear(in_features=16,out_features=32),
            nn.Dropout(0.30),
            nn.Linear(in_features=32,out_features=16),
            nn.Dropout(0.30),
            nn.ReLU(),
            nn.Linear(in_features=16,out_features=1),
            nn.Sigmoid()
            
        )
    def forward(self,x):
        return (self.layer1(x))
```

```python
model=DiseasePredict()
```

```python
loss_fn=nn.BCELoss()
optimizer=torch.optim.Adam(lr=0.001,params=model.parameters())
```

```python
epochs=10
for epoch in range(epochs):
    model.train()
    y_preds=model.forward(x_train)
    print(y_preds)
    loss=loss_fn(y_preds,y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print(loss)
```

```python
with torch.inference_mode():
    y_res=model.forward(x_test)
    for i in range(len(y_res)):
        y_res[i]=round(float(y_res[i]))
    print(y_res)
```
