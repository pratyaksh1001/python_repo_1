```python
class MultiClassifier(nn.Module):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)

        self.layer1=nn.Sequential(
            nn.Conv2d(in_channels=1,out_channels=10,kernel_size=1,padding=1),
            nn.ReLU(),
            nn.Flatten(),
            nn.Linear(9000,100),
            nn.ReLU(),
            nn.Linear(100,10),
            nn.Linear(10,10),
            nn.LogSoftmax(),
        )
    def forward(self,x):
        return (self.layer1(x)).argmax()
```

```python
model1=MultiClassifier()#.to(gpu)
summary(model1)
```

```python
loss_func=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(lr=0.01,params=model1.parameters())
```

```python
epochs=1
for epoch in range(epochs):
    y_res=model1.forward(images)
    print(y_res)
    loss=loss_func(y_res,labels)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print(loss)
```
