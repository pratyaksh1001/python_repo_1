```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("car-sales-extended.csv")
df
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
categories=["Make","Colour","Doors"]
```

```python
onehot=OneHotEncoder()
```

```python
transformer=ColumnTransformer([("one_hot",onehot,categories)],remainder="passthrough")
```

```python
x=df.drop("Price",axis=1)
```

```python
y=df["Price"]
```

```python
transformed_x=transformer.fit_transform(x)
```

```python
transformed_x
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(transformed_x,y,test_size=0.1)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model2=LinearRegression()
```

```python
model2.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```
