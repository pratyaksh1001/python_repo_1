```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import PIL
import cv2
```

```python
dataset_url="https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
```

```python
data=tf.keras.utils.get_file("flower_photos",origin=dataset_url,cache_dir=".",untar=True)
```

```python
flower_images={
    "roses":list(data_dir.glob("roses/*")),
    "daisy":list(data_dir.glob("daisy/*")),
    "dandelion":list(data_dir.glob("dandelion/*")),
    "sunflower":list(data_dir.glob("sunflower/*")),
    "tulips":list(data_dir.glob("tulips/*")),
}
```

```python
flower_images["roses"]
```

```python
flower_label={
    "roses":0,
    "daisy":1,
    "dandelion":2,
    "sunflower":3,
    "tulips":4
}
```

```python
img=cv2.imread(flower_images["roses"][0])
```

```python
img.shape
```

```python
img=cv2.resize(img,(180,180))
```

```python
img
```

```python
img.shape
```

```python
x,y=[],[]
```

```python
for label,images in flower_images.items():
    print(label)
    for image in images:
        x.append(cv2.resize(cv2.imread(image),(180,180)))
        y.append(flower_label[label])
```

```python
y
```

```python
y=np.array(y)
```

```python
len(x[1000])
```

```python
x=np.array(x)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
model=tf.keras.Sequential([
    tf.keras.layers.Conv2D(16,3,padding="same",activation="relu"),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(32,3,padding="same",activation="relu"),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(64,3,padding="same",activation="relu"),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(128,3,padding="same",activation="relu"),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(256,3,padding="same",activation="relu"),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(128,3,padding="same",activation="relu"),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128,activation="relu"),
    tf.keras.layers.Dense(5,activation="softmax")
])
```

```python
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=["accuracy"],
              optimizer="adam")
```

```python
model.fit(x_train,y_train,epochs=5)
```

```python
model.evaluate(x_test,y_test)
```
