```python
import datetime as dt
```

```python
date=df["Date"].tolist()
date
```

```python
year=[]
month=[]
day=[]
```

```python
for i in date:
    t=i.split("-")
    year.append(int(t[0]))
    month.append(int(t[1]))
    day.append(int(t[2]))
```

```python
year
```

```python
df["year"]=year
df["month"]=month
df["day"]=day
```

```python
df
```

```python
df=df.drop(["Date"],axis=1)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
test=df.drop(["High","Low","Open","Adj Close"],axis=1)
```

```python
y=test["Close"].to_numpy()
x=test.drop(["Close"],axis=1)
x=x.to_numpy()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
import sklearn as sk
```

```python
model=sk.ensemble.RandomForestRegressor(n_estimators=10)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
