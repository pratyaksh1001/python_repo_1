```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import sklearn
```

```python
import keras
```

```python
import tensorflow as tf
```

```python
df=pd.read_csv("car_prices.csv")
```

```python
df
```

```python
df
```

```python
df=df.dropna()
```

```python
len(df["model"].unique())
```

```python
df_final=df.drop(["seller","saledate","model",'trim',"vin"],axis=1)
```

```python
df_final
```

```python
df_final=pd.get_dummies(df_final)
```

```python
df_final
```

```python
df_final.replace({True:1,False:0},inplace=True)
```

```python
df_final
```

```python
model=keras.Sequential([
    keras.layers.Dense(215,activation=tf.nn.relu,input_shape=(215,)),
    keras.layers.Dense(215,activation=tf.nn.relu),
    keras.layers.Dense(215,activation=tf.nn.relu),
    keras.layers.Dense(1,activation=tf.nn.sigmoid)
])
```

```python
x=df_final.drop("sellingprice",axis=1)
```

```python
y=df_final["sellingprice"]
```

```python
model.compile(
    loss=keras.losses.categorical_crossentropy,
    optimizer="adam",
    metrics=["accuracy"]
)
```

```python
model.fit(x,y,epochs=5)
```
