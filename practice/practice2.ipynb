```python
import nltk
nltk.download('vader_lexicon')
import json
```

```python
data=json.load(open("news.json"))
```

```python
data=data["articles"]
data
```

```python
news_data=[]
news_dates=[]
```

```python
for i in data:
    news_data.append(i["description"])
    news_dates.append(str(i["publishedAt"]).split("T")[0])
```

```python
news_dates
```

```python
len(news_dates)
```

```python
(news_data)
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
sid=SentimentIntensityAnalyzer()
```

```python
score=0
neg=0
pos=0
comp=[]
for i in news_data:
    print(sid.polarity_scores(str(i))["compound"])
    score+=sid.polarity_scores(str(i))["compound"]
    neg+=sid.polarity_scores(str(i))["neg"]
    pos+=sid.polarity_scores(str(i))["pos"]
    comp.append(sid.polarity_scores(str(i))["compound"])
```

```python
score,pos,neg
```

```python
pos-neg
```

```python
news_dates
```

```python
comp
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(news_dates,comp,color='blue')
```

```python
d={}
for i in range(len(news_dates)):
    d[news_dates[i]]=0
```

```python
for i in range(len(news_dates)):
    d[news_dates[i]]+=comp[i]
```

```python
d
```

```python

#plt.bar(d.keys(),d.values(),color='red')
for i in d.keys():
    if d[i] < 0:
        plt.bar(i,d[i],color='red')
    else:
        plt.bar(i,d[i],color='green')
```
